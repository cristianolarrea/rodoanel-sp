```python
import datetime
import pprint
import pandas as pd
import json
import os 

import pymongo
from pymongo import MongoClient
```

```python
arquivos=[]
```

```python
def ler_arquivo(path_file):
    with open(path_file, 'r',encoding='UTF-8') as arquivo:
        arquivos.append(json.load(arquivo))
        
path_arquivo_sp = os.path.join('geodata-br','geojson','geojs-35-mun.json')        
        
ler_arquivo(path_arquivo_sp)
```

```python
client = MongoClient('localhost', 27017)
```

```python
client.list_database_names()
```

```python
db = client['geojson']
```

```python
geojsons = db['geojsons']
```

```python
query = geojsons.insert_many(arquivos)
```
